```python
eval_dataset_prompt = f"""
You are an evaluator who assesses the relevance of interview questions for a specific job posting.

Your goal is to analyze each question's alignment with the job requirements and responsibilities.

You will use a scale of 1 to 4, where each number represents the following:

1: The question is completely unrelated to the job requirements and responsibilities, or tests for skills/knowledge not mentioned in the job description.

2: The question has minimal connection to the job requirements but is too general or basic to effectively evaluate candidate qualifications.

3: The question relates to the job requirements and tests relevant skills, but could be more specific or targeted to the role's key responsibilities.

4: The question directly addresses critical job requirements, tests specific required skills, and effectively evaluates core competencies needed for the role.

Please provide your evaluation in the following format:

##Evaluation##

Key Job Requirements: (list extracted key skills and requirements from job description)

Question Analysis:
Q1, [First 100 characters of question], [Explanation of relevance/irrelevance], [Score 1-4]
Q2, [First 100 characters of question], [Explanation of relevance/irrelevance], [Score 1-4]
...
Qn, [First 100 characters of question], [Explanation of relevance/irrelevance], [Score 1-4]

##Evaluation##

Note: Each question should be evaluated on a single line with no blank lines between entries. The explanation should clearly connect the question content to specific job requirements or explain why it fails to do so.

Process all questions at a time. Give answer in a way that when I pasted the answers in a csv file and open with Excel, it is a perfectly well formatted csv with the 4 columns in all rows. Commas in the question itself should not be showing data in extra columns when opened in Excel.

Here is the job description:
Organization	Role/Position	Job Description
{job_desc}

Here are the questions (evaluate all in one go):
Subject	Company	Question
{questions}

Please do not show any samples. Assume you are doing correctly and just go ahead and process all questions in one go. I am not asking for brevity. I am constructing a dataset with this. So, always output full response to all questions.
"""
```

```python
client = anthropic.Anthropic(
    api_key=os.getenv("ANTHROPIC_API_KEY"),
)
message = client.messages.create(
    model="claude-3-5-sonnet-20241022",
    max_tokens=8192,
    messages=[
        {"role": "user", "content": eval_dataset_prompt}
    ]
)
# print(message.content)
```

```python
data = message.content[0].text.split("Question Analysis:\n")[-1].split('\n')
```

```python
data
```

```python
# Create lists to hold the extracted fields
question_ids = []
question_texts = []
reviews = []
scores = []

# Parse each string in the data list
for entry in data[:50]:
    parts = entry.split(", ")
    
    # Extracting components assuming the structure is consistent
    question_id = parts[0]  # First part is the question ID (e.g. Q1)
    question_text = parts[1].strip('"')  # Second part is the question text
    review = ", ".join(parts[2:-1])  # All middle parts (except the last) form the review
    score = int(parts[-1])  # Last part is the relevance score
    
    # Append these to our lists
    question_ids.append(question_id)
    question_texts.append(question_text)
    reviews.append(review)
    scores.append(score)

# Create a DataFrame
df4 = pd.DataFrame({
    'Question ID': question_ids,
    'Question Text': question_texts,
    'Review': reviews,
    'Score': scores
})

df4
```
